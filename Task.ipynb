```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vehicles.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df[(df['price'] >= 100) & (df['price'] <= 100000)]
```

```python
df = df.dropna(subset = ['price', 'year', 'manufacturer'])
```

```python
df.head()
```

```python
df['price'].plot(kind = 'hist', bins = 20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Number of Cars")
plt.show()
```

```python
df['manufacturer'].value_counts().head(10).plot(kind = 'bar')
plt.title("Top 10 Manufacturers")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Cars")
plt.show()
```

```python
sns.scatterplot(x = 'year', y = 'price', data = df)
plt.title("Price vs Year")
plt.show()
```

```python
sns.scatterplot(x = 'condition', y = 'price', data = df)
plt.title("Price by Condition")
plt.show()
```

```python
print("Average price by manufacturer :")
print(df.groupby('manufacturer')['price'].mean().sort_values(ascending = False).head(10))
```

```python
print("Most common car condition:")
print(df['condition'].value_counts())
```
